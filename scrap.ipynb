```python
from video_processing import *
from IPython.display import display # to display images
import matplotlib.pyplot as plt



"""
Canny does use two thresholds (upper and lower): 
- If a pixel gradient is higher than the upper threshold, the pixel is accepted as an edge. 
- If a pixel gradient value is below the lower threshold, then it is rejected.
"""
edge_uS = [50,100,150,200,250,300,350]
edge_lS = [50,100,150,200,250,300,350]

file_path = './mock_dataset/Step_1/HandWash_001_A_01_G01.mp4'

arr =[]
names = []
for edge_u in edge_uS:
    for edge_l in edge_lS:
        test_frames = video_to_3d(file_path, 10,edge=True, edge_u=edge_u, edge_l=edge_l)

        print("Processing",edge_u, edge_l,"...")
        names.append(f"edge_u: {edge_u} ; edge_l: {edge_l}")
        arr.append(test_frames[1])
arr=np.asarray(arr)

# display
for i in range(arr.shape[0]):
    frame = arr[i]
    fig, ax = plt.subplots()
    ax.imshow(frame)
    ax.set_title('Frame {}'.format(names[i]))
    ax.axis('off')
plt.show()
```
